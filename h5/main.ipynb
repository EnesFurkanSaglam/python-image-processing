```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("goruntu.png",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.title("Giriş Görüntüsü")
plt.axis('off')
plt.show()
```

```python
sift = cv2.SIFT_create()

keypoints_sift, descriptors_sift = sift.detectAndCompute(img, None)


img_sift_drawn = cv2.drawKeypoints(img, keypoints_sift, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img_sift_drawn)
plt.title("SIFT Anahtar Noktalar")
plt.axis('off')
plt.show()
```

```python
brisk = cv2.BRISK_create()

keypoints_brisk, descriptors_brisk = brisk.detectAndCompute(img, None)

img_brisk_drawn = cv2.drawKeypoints(img, keypoints_brisk, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img_brisk_drawn)
plt.title("BRISK Anahtar Noktalar")
plt.axis('off')
plt.show()
```

```python
print("SIFT Sonuçları")
print("Anahtar noktalar var mı: ", "Evet" if keypoints_sift else "Hayır")
print("Tanımlayıcılar var mı: ", "Evet" if descriptors_sift is not None else "Hayır")
print("Anahtar nokta sayısı: ", len(keypoints_sift))
print("Tanımlayıcı sayısı: ", descriptors_sift.shape[0] if descriptors_sift is not None else 0)

print("\nBRISK Sonuçları")
print("Anahtar noktalar var mı: ", "Evet" if keypoints_brisk else "Hayır")
print("Tanımlayıcılar var mı: ", "Evet" if descriptors_brisk is not None else "Hayır")
print("Anahtar nokta sayısı: ", len(keypoints_brisk))
print("Tanımlayıcı sayısı: ", descriptors_brisk.shape[0] if descriptors_brisk is not None else 0)
```
