```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.array([[90, 120, 130],
                [80, 100, 110],
                [70, 90, 100]], dtype=np.uint8)
```

```python
print("f(1,2) =", img[1,2])
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title("Gray Img (3x3)")
plt.colorbar()
plt.show()
```

```python
quantized = (img / 16).astype(np.uint8) * 16
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title("Original")
```

```python
plt.subplot(1, 2, 2)
plt.imshow(quantized, cmap='gray', vmin=0, vmax=255)
plt.title("Kuantize (4-bit)")
```
