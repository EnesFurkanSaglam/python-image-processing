```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('resim.jpg')

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.title("Original")
plt.imshow(img)
plt.axis("off")
```

```python
plt.title("RGB")
plt.imshow(rgb)
plt.axis("off")
```

```python
plt.title("HSV")
plt.imshow(hsv)
plt.axis("off")

plt.show()
```

```python
plt.title("Grayscale")
plt.imshow(gray)
plt.axis("off")

plt.show()
```
