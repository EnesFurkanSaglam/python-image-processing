```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('resim.jpg', 0)

plt.imshow(img)
plt.title("Original")
plt.axis('off')
plt.show()
```

```python
negative = 255 - img

plt.imshow(negative)
plt.title("Negative")
plt.axis('off')
plt.show()
```

```python
c = 255 / np.log(1 + np.max(img))
log_img = c * np.log(1 + img.astype(np.float32))
log_img = np.array(log_img, dtype=np.uint8)

plt.imshow(log_img)
plt.title("Log")
plt.axis('off')
plt.show()
```

```python
gamma = 0.5  # <1 parlaklaştırır
gamma_corrected = np.array(255 * ((img / 255) ** gamma), dtype='uint8')

plt.imshow(gamma_corrected)
plt.title("Gamma")
plt.axis('off')
plt.show()
```

```python
gamma = 2  # <1 parlaklaştırır
gamma_corrected = np.array(255 * ((img / 255) ** gamma), dtype='uint8')

plt.imshow(gamma_corrected)
plt.title("Gamma")
plt.axis('off')
plt.show()
```

```python
equalized = cv2.equalizeHist(img)

plt.imshow(equalized)
plt.title("Histogram Equalized")
plt.axis('off')
plt.show()
```
