```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('resim.jpg', cv2.IMREAD_GRAYSCALE)

avg_filtered = cv2.blur(img, (3, 3))

gaussian_filtered = cv2.GaussianBlur(img, (3, 3), 0)

median_filtered = cv2.medianBlur(img, 3)

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

sobel_combined = cv2.magnitude(sobel_x, sobel_y)
sobel_combined = cv2.convertScaleAbs(sobel_combined)

laplacian_filtered = cv2.Laplacian(img, cv2.CV_64F)
laplacian_filtered = cv2.convertScaleAbs(laplacian_filtered)

titles = ['Orijinal', 'Ortalama (Blur)', 'Gaussian', 'Median', 'Sobel (X+Y)', 'Laplacian']
images = [img, avg_filtered, gaussian_filtered, median_filtered, sobel_combined, laplacian_filtered]

plt.figure(figsize=(15, 8))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
