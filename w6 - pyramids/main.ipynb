```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("resim.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
layer = img.copy()
gp = [layer] 
import os
```

```python
for i in range(3):  # 3 katman
    layer = cv2.pyrDown(layer)
    gp.append(layer)
    
    plt.figure(figsize=(4, 4))
    plt.imshow(layer)
    plt.title(f"Gaussian Level {i+1}")
    plt.axis('off')
    plt.show()
```

```python
lp = []
for i in range(2, 0, -1):
    size = (gp[i-1].shape[1], gp[i-1].shape[0]) 
    gaussian_expanded = cv2.pyrUp(gp[i], dstsize=size)
    laplacian = cv2.subtract(gp[i-1], gaussian_expanded)
    lp.append(laplacian)

    laplacian_rgb = cv2.cvtColor(laplacian, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(4, 4))
    plt.imshow(laplacian_rgb)
    plt.title(f"Laplacian Level {i}")
    plt.axis('off')
    plt.show()
```
